엑셀...pivot table 기능이 있다...이것과 동일한 기능  
Pandas에서는 pivot_table() 함수를 제공  

pivot 함수 역할?

즉, DataFrame의 형태를 분석에 용이한 형태로 바꾸는 용도.

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}

df1 = DataFrame(data)

df1
```

```python
# 위치 매개변수에 따라 표가 작성
# 년도별 도시의 인구
df1.pivot_table('인구','도시','연도')
```

```python
# 키워드 매개변수 : 매개변수 순서 무시하고 명시해서 pivot 완성
df1.pivot_table('인구',columns='연도',index='도시')
```

```python
# 여러 인덱스, 컬럼을 사용하여 세분화 가능
df1.pivot_table('인구',index=['도시'],columns=['연도'])
```

```python
# 각 행, 열 값의 평균 값 : margins = True 
print(df1['인구'].mean()) # 확인
df1.pivot_table('인구',index=['도시'],columns=['연도'], margins=True)
```

```python
df1.pivot_table(['인구'], index=['연도','도시'])
```

## 실전데이터 응용

```python
tipdf = pd.read_csv('../data/tips.csv')
tipdf.head()
```

```python
'''
해당 tips 데이터의 분석의 목표는 사실상...tip비율
식사대금 대비 팁의 비율이 어떤지 알수 있는 컬럼 하나를 더 추가해보자.
tip_pct = tip / total_bill
'''
tipdf['tip_pct'] = tipdf['tip'] / tipdf['total_bill']
tipdf.tail()

# NaN 누락데이터 제거...
tipdf.dropna(how='any', inplace=True)
tipdf.tail()

# 데이터들에 대한 간략한 통계정보 출력...분포
tipdf.describe()
tipdf.info()
tipdf
```

pivot table의 인덱스 기능은 groupby 와 똑같다...즉, 세분화!!

```python
# 문제 1. tipdf에서 pivot를 사용...day컬럼으로 세분화
tipdf.pivot_table(index=['day'], aggfunc='mean').round(2) # 소수점 3번째 자리에서 반올림
```

```python
# 문제 2. tipdf 에서 모든 데이터를 성별. 흡연여부로 세분화...소수점 2자리까지만 출력
tipdf.pivot_table(index=['sex','smoker']).round(2)
```

```python
'''
pivot_table() 함수 인자값 속성의 의미

values - 데이터프레임을 채우고자 하는 값
index - 세분화, 그룹핑 하는 기준
columns - 컬럼값
'''
tipdf.pivot_table(
    values=['total_bill','tip'],
    index=['sex','day'],
    columns=['smoker']
)
```

```python
tipdf.pivot_table('tip','sex','smoker')
```

```python
# 문제 3.
# 성별에 따른 평균팁의 비율
'''
1) pivot을 사용
'''
tipdf.pivot_table(values=['tip_pct'],index=['sex']).round(2)
```

```python
'''
2) groupby()
'''
tipdf.groupby(by=['sex'])[['tip_pct']].mean().round(2)
```

```python
# 문제 4. ... groupby()로 해결하시오.
'''
각 그룹에서 가장 많은 팁과 가장 적은 팁의 차이를 보이시오.
단 사용자 정의 함수를 지정...max_min_tip 으로 정의 (agg, apply)
'''
def max_min_tip(x):
    return x.max() - x.min()

tipdf.groupby(['sex','smoker'])[['tip']].agg(max_min_tip)
```
