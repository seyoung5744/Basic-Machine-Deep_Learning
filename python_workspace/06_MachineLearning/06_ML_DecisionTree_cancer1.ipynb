```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

### 1. Data Loader

```python
cancer = datasets.load_breast_cancer()
cancer
```

### 2. Cancer Data Analysis

```python
X = cancer.data
y = cancer.target
print('Feature_names \n', cancer.feature_names)
print('*'*100)
print('Target_names \n', cancer.target_names)
print('*'*100)
print('Target_names \n', cancer.data.shape)
```

```python
# 데이터 확인

# dir(cancer)
# cancer.keys()

np.bincount(cancer.target) # target 0, target 1
cancer.target[cancer.target==0].shape
```

```python
print("Sample Count per class : \n{}".format(
    {n: v for n, v in zip (cancer.target_names,
                           np.bincount(cancer.target))}))
```

```python
for i, name in enumerate(cancer.feature_names):
    print("%02d : %s" %(i, name)) # 2자리수로 출력
```

### 3. Model Generator and Train

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
dTree = DecisionTreeClassifier()
dTree.fit(X_train, y_train)
dTree
```

### 4. Predict

```python
pred = dTree.predict(X_test)
pred
```

### 5. Model Evaluate

```python
score = accuracy_score(pred, y_test, normalize=True) # 디폴트...정확히 예측한 확률값
score = accuracy_score(pred, y_test, normalize=False) # 정확히 예측한 데이터의 건수
score
```

```python
import graphviz
from sklearn.tree import export_graphviz
export_graphviz(dTree, out_file='cancer_tree.dot',
                class_names=cancer.target_names,
                feature_names=cancer.feature_names,
                filled = True) 

with open('cancer_tree.dot', encoding='utf-8') as f:
    dot_graph = f.read()
dot = graphviz.Source(dot_graph)
display(dot)
dot.render(filename='cancer_tree.png')
```
